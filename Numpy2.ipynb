```python
import datetime
f=open("data.txt")
s=f.read()
l=eval(s)
print(type(l))
s=set()
s1=set()
for i in l:
    s.add(i['student_id'])
    s1.add(i['date'])
s1=list(s1)
s1.sort()
d={}
for i in l:
    if(i['date']==s1[0]):
        if(i['student_id'] not in list(d.keys())):
            d[i['student_id']]=list()
            d[i['student_id']].append(i['attendance'])
        else:
            d[i['student_id']].append(i['attendance'])
    if(i['date']==s1[1]):
        if(i['student_id'] not in list(d.keys())):
            d[i['student_id']]=list()
            d[i['student_id']].append(i['attendance'])
        else:
            d[i['student_id']].append(i['attendance'])
    if(i['date']==s1[2]):
        if(i['student_id'] not in list(d.keys())):
            d[i['student_id']]=list()
            d[i['student_id']].append(i['attendance'])
        else:
            d[i['student_id']].append(i['attendance'])
for i in d.keys():
    if(len(d[i])<6):
        d[i].insert(2,"NA")
    print(d[i])
```

```python
import re
reg = '^\w+([\.-]?\w+)*@\w+([\.-]?\w+)*(\.\w{2,3})+$'
def check(email):  
    if(re.search(reg,email)):  
        print("Valid Email")            
    else:  
        print("Invalid Email")  
       
email = input("enter the Email Id: ")
check(email)
```

```python
import re
reg = '\\w{3}\\w{3}\\w{4}'
def check(phone):  
    if(re.search(reg,phone)):  
        print("Valid Phone number")            
    else:  
        print("Invalid Phone number")  
       
phone = input("enter the Phone number: ")
check(phone)
```

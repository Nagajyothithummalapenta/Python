###  Text Processing :Stemming,Stop-words removal and Lemmatization
- Begin by removing html tags
- Remove any punctuation or set of any special characters
- Check if the word is made up of english letters and is not alpha-numeric
- Convert the word to lowercase
- Remove stopwords
- Finally snowball stemming the word

```python
import nltk # natural language tool kit
nltk.download('stopwords')
import re as stdre
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
stop=set(stopwords.words('english')) # set of stopwords
print(stop)
```

```python
sno=nltk.SnowballStemmer('english')
print(sno.stem('computers'))
```

```python
def cleanhtml(sentence):
    cleanr=stdre.compile('<.*?>')
    cleantext=stdre.sub(cleanr,' ',sentence)
    return cleantext
```

```python
cleanhtml(data.Message[2297])
```

```python
def cleanpunc(sentence):
    cleaned=stdre.sub('[?|!|\'|"|#]',r'',sentence)
    cleaned=stdre.sub('[.|,|)|(|\|/)]',r' ',cleaned)
    return cleaned
```

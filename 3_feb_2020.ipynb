```python
np.random.seed(10)
a=np.random.randint(10,200,50)
a
```

create a new array
a>50 and a<150 - good

a<50- bad

a>150-Excellent

```python
l=[]
for i in a:
    if i>=50 and i<=150:
        l.append('good')
    elif(i<50):
        l.append('bad')
    elif(i>150):
        l.append('Excellent')
print(l)
```

```python
def number_cat(n):
    if(n>=50)and(n<=150):
        return 'good'
    elif(n<50):
        return 'bad'
    elif(n>150):
        return 'Excellent'
number_cat(200)
```

```python
number_cat_vect=np.vectorize(number_cat)
```

```python
number_cat_vect(a)
```

```python
import pandas as pd
cat=pd.Series(number_cat_vect(a))
cat
```

```python
cat.value_counts()
```

```python
y_values=cat.value_counts()
x_values=cat.value_counts().index
plt.bar(x_values,y_values)
```

```python
plt.bar(x_values,y_values,color='rgb')
plt.legend("A bar Graph")
```

```python
x=np.arange(10)
plt.plot(x,x*3,'r',label='Line')
plt.plot(x,x**3,'b',label='cube')
plt.legend(loc='upper right')
```

```python
plt.bar(x_values,y_values,color='grb')
colors={'good':'green','bad':'red','excellent':'blue'}
labels=list(colors.keys())
handles=[plt.Rectangle((0,0),1,1,color=colors[label]) for label in labels]
plt.legend(handles,labels)
```
